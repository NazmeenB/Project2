```python
def handle_outliers(df, target_column):
    for col in df:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Cap outliers instead of removing them
        df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
        df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    return df
```

```python
df = handle_outliers(df, 'category')
```

```python
for column in df.columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot of {column}")
    plt.show()
```

```python
numerical_columns = df.select_dtypes(include='number').columns

# Function to handle outliers using IQR
def handle_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# Adjust the number of rows and columns for subplots
n_cols = 5  # Number of plots per line
n_rows = (len(numerical_columns) + n_cols - 1) // n_cols  # Calculate rows based on the number of columns

# Create a figure with subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 5))
axes = axes.flatten()  # Flatten to easily index

# Loop through numerical columns and plot each boxplot after handling IQR
for i, col in enumerate(numerical_columns):
    cleaned_data = handle_outliers_iqr(df, col)
    sns.boxplot(data=cleaned_data, x=col, ax=axes[i])
    axes[i].set_title(f'Boxplot for {col} (IQR)', fontsize=12)

# Remove any empty subplots if the grid is larger than the number of plots
for j in range(len(numerical_columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
x = df.drop(columns='category')
y = df['category']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

## 1. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
#creating a decision tree model from entropy method. 
model = DecisionTreeClassifier(criterion="entropy",max_depth=5, random_state= 42)
```

```python
model.fit(x_train, y_train)
```

```python
a = model.score(x_test , y_test)  
print(f"Accuracy score for Decision Tree: {a:.3f}")
```
